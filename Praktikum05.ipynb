**A. Scatter Plot**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Membuat scatter plot
plt.scatter(suhu, penjualan, color='blue')
plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.show()
```

**B. Kustomisasi Scatter Plot**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Kustomisasi scatter plot
plt.scatter(
    suhu,
    penjualan,
    color='orange',
    s=100,
    edgecolor='black',
    alpha=0.7
)

plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.grid(True)

plt.show()
```

**C. Menambahkan Scatter Plot**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Data tambahan berdasarkan kategori hari
penjualan_kerja = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_akhir_pekan = [60, 70, 80, 100, 110, 120, 140, 160, 200]

# Scatter plot multiple
plt.scatter(
    suhu,
    penjualan_kerja,
    color='green',
    label='Hari Kerja',
    s=80
)

plt.scatter(
    suhu,
    penjualan_akhir_pekan,
    color='purple',
    label='Akhir Pekan',
    s=80
)

plt.title('Penjualan Es Krim Berdasarkan Hari')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.legend()

plt.show()
```

**D. Scatter Plot untuk Melihat Hubungan Data dengan Beberapa Variabel**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Data tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [200, 220, 240, 260, 280, 300, 320, 340, 360]

# Scatter plot dengan ukuran dan warna bervariasi
plt.scatter(
    suhu,
    penjualan,
    c=kelembapan,
    s=ukuran_titik,
    cmap='coolwarm',
    alpha=0.6
)

plt.title('Hubungan Penjualan, Suhu, dan Kelembapan')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.colorbar(label='Kelembapan (%)')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# 1. Data Dummy
data = {
    'Suhu': [20, 22, 24, 26, 28, 30, 32, 34, 36],
    'Kelembapan': [60, 65, 70, 75, 80, 85, 90, 95, 100],
    'Penjualan': [50, 60, 70, 90, 100, 110, 130, 150, 180]
}
df = pd.DataFrame(data)

# 2. Membuat Figure
fig = plt.figure(figsize=(10, 7))

# 3. Menambahkan Subplot 3D
ax = fig.add_subplot(111, projection='3d')

# 4. Membuat Scatter Plot 3D
ax.scatter(df['Suhu'], df['Kelembapan'], df['Penjualan'], c='red', s=100, marker='o')

# 5. Label Sumbu
ax.set_xlabel('Suhu (°C)')
ax.set_ylabel('Kelembapan (%)')
ax.set_zlabel('Penjualan Es Krim')
ax.set_title('3D Scatter Plot: Suhu, Kelembapan, dan Penjualan')

# 6. Tampilkan
plt.show()
```
